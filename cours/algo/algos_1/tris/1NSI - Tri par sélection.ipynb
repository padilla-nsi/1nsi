Le tri par sélection parcourt le tableau de la gauche vers la droite, en maintenant sur la gauche une partie déjà triée et à sa place définitive

À chaque étape on cherche le plus petit élement dans la partie droite non triée, puis on l'échange avec l'élément le plus à gauche de la partie non triée.

Ainsi la première étape va déterminer le plus petit élément et le placer tout à gauche du tableau. Puis la deuxième étape va déterminer le deuxième plus petit élément et le placer dans la deuxième case du tableau, et ainsi de suite.

Par construction, la partie gauche déjà triée ne contient que des éléments inférieurs ou égaux à ceux de la partie droite restant à trier.
