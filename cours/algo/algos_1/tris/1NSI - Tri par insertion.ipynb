Comme le tri par sélection, le tri par insertion parcourt un tableau de la gauche vers la droite et maintenant sur la gauche une partie déjà triée.

Mais au lieu de chercher/sélectionner/permuter la plus petite valeur, le tri par insertion va *insérer* la première valeur non encore triée (c'est-à-dire la valeur la plus à gauche de la partie droite non triée) dans la partie gauche déjà triée.

Pour cela, on va décaler d'une case vers la droite tous les éléments déjà triés qui sont plus grand que la valeur à insérer, puis déposer cette dernière dans la case ainsi libérée

#### Algorithme

Visualiser l'animation *Insertion sort* du site suivant :

#### Implémentation

Comme pour le tri par sélection, créer d'abord une fonction `permute(tab: list, i: int, j: int)` qui permute les éléments `tab[i]` et `tab[j]`.

Implémenter une fonction `tri_par_insertion(tab)` qui trie un tableau en appliquant l'algorithme du tri par insertion.
